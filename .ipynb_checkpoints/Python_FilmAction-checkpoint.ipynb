```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re



#   DF
filmIDDF = pd.DataFrame()

#1     DF
filmIDDF_temp = pd.DataFrame()
filmIDDF_temp = filmIDDF_temp.append(
    {"Title":"",
     "Year" : "",
     "Genre" : "",
     "Runtime" : "",
     "Certificate" : "",
     "Rating" : "",
     "Votes" : "",
     "Metascore" : "",
     "Director" : "",
     "Cast": ""},
    ignore_index = True)

url_part1 = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="
url_part2="&ref_=adv_nxt"

for j in range(1, 1400,50):
    url = url_part1 + str(j) + url_part2
    res = requests.get(url) # urllib,request.urlopen(~~)
    soup = bs(res.text, "html.parser")

    num = len(soup.find_all("h3"))-1
    #print(num)
    print(url)
    print(num)

    for i in range(0,num):
        temp = soup.find_all("div", class_="lister-item mode-advanced")[i]
        
        # ????1!!!!---> temp   i     ....
        # 
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Title")] = temp.find("h3").find("a").text
        # 
        Year = temp.find("h3").find_all("span")[1].text.strip()
        Year = Year.replace("(","")
        Year = Year.replace(")","")
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Year")] = Year
        # 
        Genre = temp.find("div", class_="lister-item-content").find("p", class_="text-muted").find("span", class_="genre").text
        Genre = Genre.replace("\n","")
        Genre = Genre.replace("            ","")
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Genre")] = Genre
        # 
        Runtime = temp.find("div", class_="lister-item-content").find("p", class_="text-muted").find("span", class_="runtime").text
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Runtime")] = Runtime
        # 
        Cert = temp.find("div", class_="lister-item-content").find("p", class_="text-muted").find("span", class_="certificate")
        if Cert != []:
            filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Certificate")] = Cert
        else:
            filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Certificate")] = ""
          # Cert = soup.find_all("div", class_="lister-item-content")[i].find("p", class_="text-muted").find("span", class_="certificate").text
          # filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("")] = Cert
        # 
        Ratings = temp.find("div", class_="inline-block ratings-imdb-rating").find("strong").text
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Rating")] = Ratings
        # 
        Votes = temp.find("", class_="sort-num_votes-visible").find_all("span")[1].text
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Votes")] = Votes
        # 
        Metaspace = temp.find_all("div", class_="inline-block ratings-metascore")
        if Metaspace != []:
            Metascore = temp.find("div", class_="inline-block ratings-metascore").find("span").text
            Metascore = Metascore.replace("        ","")
            filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Metascore")] = Metascore
        else:
            filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Metascore")] = "None"
        #
        Director = temp.find_all("p")[2].find("a").text
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Director")] = Director
        #
        Cast = temp.find_all("p")[2].find_all("a")[1].text
        filmIDDF_temp.iat[0, filmIDDF_temp.columns.get_loc("Cast")] = Cast
        


        filmIDDF=filmIDDF.append(filmIDDF_temp, ignore_index=True)

# if 
# filmIDDF["Genre_1"] = filmIDDF["Genre"].apply(lambda x: str(x).split(',')[0]) 
# filmIDDF["Genre_2"] = filmIDDF["Genre"].apply(lambda x: str(x).split(',')[1])

filmIDDF["Genre_1"] = filmIDDF["Genre"].apply(lambda x: str(x).split(',')[0]) 
#filmIDDF["Genre_2"] = filmIDDF["Genre"].apply(lambda x: str(x).split(',')[1]) 


filmIDDF.drop_duplicates()

columnList = ["Title", "Year", "Certificate", "Runtime", "Genre", "Metascore","Votes", "Rating", "Director","Cast","Genre_1"]
filmIDDF = filmIDDF[columnList]

filmIDDF
```

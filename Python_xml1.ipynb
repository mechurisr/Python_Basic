```python
#모든 정보를 담을 DF
movieIDDF = pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append(
    {"movieCd":"",
     "movieNm" : "",
     "movieNmEn" : "",
     "openDt" : "",
     "directorpeoplenm" : ""},
    ignore_index = True)

num = len(soup.find_all("movie"))

for i in range(0,num):
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieCd")] = soup.find_all("movie")[i].find("moviecd").text
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieNm")] = soup.find_all("movie")[i].find("movienm").text
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieNmEn")] = soup.find_all("movie")[i].find("movienmen").text
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("openDt")] = soup.find_all("movie")[i].find("opendt").text
    

    #print(str(i)+"_번째 영화 코드: ", movieIdData["movieListResult"]["movieList"][i]["directors"]=[])
    #감독에 대한 정보가 뭐라도 들어있는 경우....
    if soup.find_all("movie")[i].find("directors").find_all("peoplenm") !=[]:
        #soup.find_all("movie")[i].find("directors").text !=""
         movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("directorpeoplenm")] = soup.find_all("movie")[i].find("directors").find_all("peoplenm")[0].text
    else:
            pass
        
    

    
    # 1. movieIDDF_temp에 i번째 영화의 정보들을 담기....(코드, 이름, 영문, 개봉일, 감독)
    # 2. movieIDDF_temp에 있는 i번째 영화에 대한 정보를 movieIDDF에 추가
    # ---------> append
    movieIDDF=movieIDDF.append(movieIDDF_temp, ignore_index=True)
    # 3. movieIDDF_temp에 담겨져 있는 정보들을 초기화하는 부분
    movieIDDF_temp.iat[0, temp_movcd_idx] = ""
    movieIDDF_temp.iat[0, temp_movNm_idx] = ""
    movieIDDF_temp.iat[0, temp_movNmE_idx] = ""
    movieIDDF_temp.iat[0, temp_movdt_idx] = ""
    movieIDDF_temp.iat[0, temp_movdir_idx] = ""
```

```python
#M1) 
# Real Code : iat 사용!!!!
# --> 그러면 만든 row, column 의 정수 정보로 통일해야 함!!!!


# 여기서는 이미 DataFrame이 만들어져 있기 때문에 다음과 같이 간편히 컬럼을 추가하고 초기화를 할 수 있음!
# 위에서 맨 처음에 아무 정보도 없는 DataFrame 을 만드는 과정과 다름을 이해할 것!!!!
# 이 부분에 대한 실수로 인해서 초기화 작업에 혼동이 될 수 있음!!!!!!
movieIDDF["type"] = ""
movieIDDF["ActorNum"] = ""
movieIDDF["cast"] = ""
movieIDDF["cast_peopleNm"] = ""


i_type = movieIDDF.columns.get_loc("type")
i_acn = movieIDDF.columns.get_loc("ActorNum")
i_cast = movieIDDF.columns.get_loc("cast")
i_castpeo = movieIDDF.columns.get_loc("cast_peopleNm")



dataLen = len(movieIDDF)
for i in range(0, dataLen):
    tempURL_Part1 = """http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?key="""
    keyNum = """80b1793712960b5dfe1c3386942f37d9"""
    tempURL_Part2 = """&movieCd="""
    movieId = str(movieIDDF.index[i])
    tempURL = tempURL_Part1 + keyNum + tempURL_Part2 + movieId
    
    # For Check URL
    resTest = urllib.request.urlopen(tempURL)
    soupTest = BeautifulSoup(resTest,"html.parser")

    movieIDDF.iat[i,i_type] = soupTest.find_all("typenm")[0].text
    movieIDDF.iat[i,i_acn] = len(soupTest.find_all("actor"))
    
    if len(soupTest.find_all("actor")) <1:
        movieIDDF.iat[i,i_cast] = ""
    else:
        movieIDDF.iat[i,i_cast] = soupTest.find_all("actor")[0].find("cast").text
        movieIDDF.iat[i,i_castpeo] = soupTest.find_all("actor")[0].find("peoplenm").text
movieIDDF
```

```python
import pandas as pd
import numpy as np
```

```python
## pivot ##
data = pd.read_excel("C:\\Users\\acorn-519\\Desktop\\sampledata\\11_sales-funnel.xlsx")
data.head()
```

```python
data.head(10)
```

```python
data["Status"] = data["Status"].astype("category")

data["Status"].cat.set_categories(
    ["won", "pending", "presented", "declined"],
    inplace = True)

data.head()
```

```python
data.info()
```

```python
# Name을 가로줄 기준으로 하고, 데이터를 보자 --> 숫자인 컬럼들만 들어온다
pd.pivot_table(data,index = ["Name"])
```

```python
pd.pivot_table(data,index = ["Name","Rep","Manager"])
```

```python
pd.pivot_table(data,
               index = ["Manager","Rep"], # 가로줄 세팅
               values = ["Price", "Quantity"], # 세로줄 세팅
               aggfunc = [np.sum, np.mean], #각 칸을 무엇으로 채울지 결정..여기서는 합으로
               fill_value = 0,
               margins = True)
               
```

```python
#Price 에는 aggfunc : np.sum. Quantity 에는 aggfunc : len
table = pd.pivot_table(data,
               index = ["Manager","Status"], # 가로줄 세팅
               values = ["Price", "Quantity"], # 세로줄 세팅
               aggfunc = {"Price" : np.sum, "Quantity":len}, #각 칸을 무엇으로 채울지 결정..여기서는 합으로
               columns = ["Product"],
               fill_value = 0,
               margins = True)
table
```

```python
table.query('Status == ["won","pending"]')
```

```python
pd.crosstab([data["Manager"],data["Rep"]],[data["Price"], data["Quantity"]],
           margins = True)
```

# 수정 : 누락 없이 들어가게 하는 방식[append / concat ]

```python
import time
```

```python
# APPEND USE

# ID 테이블 만들기 : 참고 ---> 수정

t1 = time.time()

movieIDDF = pd.DataFrame()
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append({"movieCd":"", "movieNm" : "", "movieNmEn" : "", "openDt":"","directorpeopleNm":""},ignore_index=True)


# 주의!!!) pandas가 업데이트 되면서 새로운 컬럼을 자동 생성을 하면서 가지를 않는다;;;;
# 과거에는 이렇게 set_value로 처리가 간단히 가능했음;;; 최근 업데이트로 인해서는 이러한 간단한 방법이 되지 않음.ㅠㅠ
# 넣을 데이터를 데이터 프레임으로 만들고 concate, append로 처리해야함!!
num = len(movieIdData["movieListResult"]["movieList"])
print(num)
for i in range(0, num):
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieCd")] = movieIdData["movieListResult"]["movieList"][i]["movieCd"]
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieNm")] = movieIdData["movieListResult"]["movieList"][i]["movieNm"]
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("movieNmEn")] = movieIdData["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("openDt")] = movieIdData["movieListResult"]["movieList"][i]["openDt"]
    
    if movieIdData["movieListResult"]["movieList"][i]["directors"] != []:
        movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("directorpeopleNm")] = movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else:
        movieIDDF_temp.iat[0, movieIDDF_temp.columns.get_loc("directorpeopleNm")] = ""

    # 합치기
    movieIDDF = movieIDDF.append(movieIDDF_temp,ignore_index=True)
    
    # 다시 초기화
    for j in range(0, len(movieIDDF_temp.columns)):
        movieIDDF_temp.iat[0,j]=""
t2 = time.time()
print(t2-t1)
movieIDDF    
```

* Q) 위의 json의 파일을 가지고 movieIDDF 라는 DataFrame을 만들고 나서, openDt의 정보를 활용하여 year, month, day의 컬럼의 값을 생성하기.[힌트 : 앞에서 여러 번 한 apply/lambda적용, 문자열의 일부 추출하는 방법 생각해보기!!!)

```python
# openDt에서 앞에 4자리만 이용해서 year 컬럼을 만들고, month, day칼럼 만들기..
# 문자열과 apply등의 적용 및 index와 관련된 혼동을 확인하기 위한 예제
movieIDDF["year"] = movieIDDF["openDt"].apply(lambda x: str(x)[:4]) 
movieIDDF["month"] = movieIDDF["openDt"].apply(lambda x: str(x)[4:6])
movieIDDF["day"] = movieIDDF["openDt"].apply(lambda x: str(x)[6:])
movieIDDF
```

* 날짜 순으로 정렬을 하고, index를 movieCd로 하고, 
*  컬럼 순서를 openDt 	year 	month 	day  movieNm 	movieNmEn directorpeopleNm 로 정리


```python
# 참고) 이미 set_index  한 이후에, 다시 하게 되면 에러가 발생한다!!!
movieIDDF = movieIDDF.sort_values(by ="openDt")
movieIDDF = movieIDDF.set_index("movieCd")

# 원하는 순서대로 컬럼 변경
movieIDDF = movieIDDF[["openDt","year","month","day", "movieNm","movieNmEn","directorpeopleNm"]]

movieIDDF
```

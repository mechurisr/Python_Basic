```python
import urllib.request
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

wageURL = "http://apis.data.go.kr/B551982/openApiFullTimeTotal/openXmlFullTimeTotal?serviceKey=BClVDLLEomkCLzflApWLzmRn%2B3B7krZSu7fVnKiaspmbTbmNm4kxOxh3LUeoRc136A5KnVXgV4H%2FJSqAg7EULg%3D%3D&type=xml&numOfRows=10&pageNo=1&ac_year=2016"
res = urllib.request.urlopen(wageURL)
wageSoup = BeautifulSoup(res, "html.parser")


#모든 정보를 담을 DF
wageIDDF = pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
wageIDDF_temp = pd.DataFrame()
wageIDDF_temp = wageIDDF_temp.append(
    {"기관명":"",
     "기본급" : "",
     "복리후생" : "",
     "수당" : "",
     "직원수" : "",
     "평균근속년수" : "",
     "평균임금" : ""},
    ignore_index = True)

num = len(wageSoup.find_all("ent_name"))

for i in range(0,num):
    
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("기관명")] = wageSoup.find_all("item")[i].find("ent_name").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("기본급")] = wageSoup.find_all("item")[i].find("t_gi").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("직원수")] = wageSoup.find_all("item")[i].find("t_staff").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("평균근속년수")] = wageSoup.find_all("item")[i].find("t_wyear_wmonth").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("평균임금")] = wageSoup.find_all("item")[i].find("t_one_ave_pay").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("복리후생")] = wageSoup.find_all("item")[i].find("t_pay_etc").text.strip()
    wageIDDF_temp.iat[0, wageIDDF_temp.columns.get_loc("수당")] = wageSoup.find_all("item")[i].find("t_jesu").text.strip()
    
    wageIDDF=wageIDDF.append(wageIDDF_temp, ignore_index=True)
    
    for j in range(0, len(wageIDDF_temp.columns)):
        wageIDDF_temp.iat[0, j]= ""
        

wageIDDF = wageIDDF.astype({'기본급':'float64', '복리후생':'float64', '수당':'float64', '직원수':'float64', '평균임금':'float64'})

columnList = ["기관명", "기본급", "복리후생", "수당", "직원수", "평균근속년수", "평균임금"]
wageIDDF = wageIDDF[columnList]

wageIDDF
```

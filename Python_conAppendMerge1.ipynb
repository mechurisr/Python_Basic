```python
#01 DataSet
baseball_player = pd.DataFrame (
    {
        "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993, 2001, 2001, 2010, None]
    }
)
baseball_team = pd.DataFrame (
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    }
)
print(baseball_player)
print(baseball_team)
```

```python
pd.concat([baseball_player, baseball_team], axis = 1)
```

```python
pd.merge(baseball_player,baseball_team,
         how = "inner",
         left_on = "team_No",
         right_on = "team_id")
#두 테이블에 공통된 것만 출력(아닐경우 누락)
```

```python
pd.merge(baseball_player, baseball_team, 
        how = "left",
        left_on = "team_No",
        right_on = "team_id")
```

```python
pd.merge(baseball_player, baseball_team, 
        how = "right",
        left_on = "team_No",
        right_on = "team_id")
```

```python
temp =pd.merge(baseball_player, baseball_team, 
        how = "left",
        left_on = "team_No",
        right_on = "team_id")
temp
```

```python
temp[temp["team_id"].isnull()]
#boolean 조건
```

```python
temp =pd.merge(baseball_player, baseball_team, 
        how = "right",
        left_on = "team_No",
        right_on = "team_id")
temp
```

```python
temp[temp["team_No"].isnull()]
#boolean 조건
```

```python
#모두포함
temp =pd.merge(baseball_player, baseball_team, 
        how = "outer",
        left_on = "team_No",
        right_on = "team_id")
temp
```

```python
temp[temp["team_No"].isnull() | temp["team_id"].isnull()]
```

```python
baseball_player = pd.DataFrame (
    {
        "player_Name":["권혁","이범호","필","박병호","양준혁"],
        "team_No":[1993, 2001, 2001, 2010, None]
    },
    index = ["p_01","p_02","p_03","p_04","p_05"]
)
baseball_player.index.name = "player_id"
#player는 기준이 컬럼 : team_No, 팀 정보는 기준이 가로 인덱스에 존재
baseball_team = pd.DataFrame (
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
    },
        index =[1993,1999,2001,2010]
)
```

```python
# inner join
pd.merge(baseball_player, baseball_team,
        how = "inner",
        left_on = "team_No",
        right_index = True)
```

```python
pd.merge(baseball_player, baseball_team,
        how = "left",
        left_on = "team_No",
        right_index = True)
```

```python
pd.merge(baseball_player, baseball_team,
        how = "right",
        left_on = "team_No",
        right_index = True)
```

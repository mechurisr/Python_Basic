```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re

date = "2017.03.28"
url_part1 = """http://dart.fss.or.kr/dsac001/mainAll.do?selectDate="""
url_part2 = """&sort=&series=&mdayCnt=0&currentPage="""

url = url_part1 + date + url_part2

res = requests.get(url) # urllib,request.urlopen(~~)
soup = bs(res.text, "html.parser")


#모든 정보를 담을 DF
dartIDDF = pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
dartIDDF_temp = pd.DataFrame()
dartIDDF_temp = dartIDDF_temp.append(
    {"Cat":"",
     "comName" : "",
     "coID" : "",
     "Content" : "",
     "rcpNo" : "",
     "pubTime" : "",
     "pubDate" : "",
     "reqDate" : ""},
    ignore_index = True)

num = len(soup.find_all("tr"))
print(num)

for i in range(1,101):
    #시장종류
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("Cat")] = soup.find_all("tr")[i].find_all("td")[1].find("img").get("title")
    #회사이름
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("comName")] = soup.find_all("tr")[i].find("a").text.strip()
    #회사ID
    testStr = soup.find_all("tr")[i].find("a").get("href")
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("coID")] = re.findall(r"[0-9]{8}", testStr)[0]
    #공시 내용/제목
    Content = soup.find_all("tr")[i].find_all("td")[2].text.strip()
    Content = Content.replace("\r","")
    Content = Content.replace("\n","")
    Content = Content.replace("\t","")
    Content = Content.replace(" ","")
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("Content")] = Content
    #해당 공시 고유값
    rcpNo = soup.find_all("tr")[i].find_all("td")[2].find("a").get("onclick")
    rcpNo = re.findall(r"[0-9]{14}", rcpNo)[0]
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("rcpNo")] = rcpNo
    #공시 제출시간
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("pubTime")] = soup.find_all("tr")[i].find("td").text.strip()
    #공시 날짜
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("pubDate")] = date
    #공시 요청날짜
    dartIDDF_temp.iat[0, dartIDDF_temp.columns.get_loc("reqDate")] = soup.find_all("tr")[i].find_all("td")[4].text

         
    dartIDDF=dartIDDF.append(dartIDDF_temp, ignore_index=True)

dartIDDF.head(101)
```

# 데이터 합치기

```python
##################### 데이터 합치기 ###########
import pandas as pd
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                      index=[4, 5, 6, 7])
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'CC': ['C4', 'C5', 'C6', 'C7'],
                     'DD': ['D4', 'D5', 'D6', 'D7']},
                      index=[ 2, 3,5,7])
print(df1)
print(df2)
```

```python
pd.concat ([df1, df2], axis=1)
```

```python
# 원하는 컬럼만 선택한다고 해도, pandas는 기본적으로 중복된 것은 다 모두 출력이 기본이었음!!!!!
pd.concat ([df1, df2], axis=1).loc[:,["A","B","CC"]]
```

```python
pd.concat ([df1, df2], axis=1).iloc[:,[0,1,6,7]]
```

```python
# 양쪽에서 모두 완전한 것만 추리기!!!!
pd.concat ([df1, df2], axis=1, join="inner")
```

```python
pd.concat ([df1, df2], axis=1).dropna()
```

```python
# df2의 인덱스를 기준으로 정리하고 싶을 때!!

# ----> 버전업이 되면서 이 기능이 사라지게 됨;;;;;;
pd.concat ([df1, df2], axis=1, join_axes=[df2.index])
```

```python
pd.concat ([df1, df2], axis=1).reindex(df2.index)
```

```python
pd.concat ([df1, df2], axis=1).reindex(df1.index)
```

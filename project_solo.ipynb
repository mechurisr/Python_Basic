```python
# 여러 파일이 있는 경로에 *.csv 를 하나의 list로
file_list = glob.glob(animation_path)+glob.glob(book_path)+\
             glob.glob(cartoon_path)+glob.glob(contents_path)+\
             glob.glob(entertainment_path)+glob.glob(gami_path)+\
             glob.glob(movie_path)+glob.glob(music_path)
```

```python
# vars() => 문자열을 변수명으로 사용하게 해주는 함수
for i in range(0,len(file_list)):
    temp = os.path.basename(file_list[i])
    temp = os.path.splitext(temp)[0]
    
    vars()[str(temp)] = pd.read_csv(file_list[i], encoding="utf-16", sep="\t", skiprows=1).set_index("표측1")
```

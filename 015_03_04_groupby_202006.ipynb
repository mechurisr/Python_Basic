```python
# Method2)
data_path ="sampledata/10_phone_data.csv"
data = pd.read_csv(data_path)
dates = data["date"]

# 이 부분은 교제 시계열 부분 참고!!398 페이지
# pd.datetime.strptime(dates[0], "%y/%m/%d %H:%M")
dateparse = lambda dates: [pd.datetime.strptime(d, '%y/%m/%d %H:%M') for d in dates]

# 아래와 같이 하면 안 되는 데이터가 있어서...ValueError: time data '15/10/14 06:58' does not match format '%y/%m/%d %H:%M:%S'
#dateparse = lambda dates: [pd.datetime.strptime(d, '%y/%m/%d %H:%M:%S') for d in dates]

data = pd.read_csv(data_path, parse_dates=['date'], date_parser=dateparse)
data.head()
```

```python
# month로 묶어 보자 ---> 바로 보이지가 않음;;;;;
data.groupby(['month'])
```

```python
# 묶은 것의 키를 보자
data.groupby(['month']).groups.keys()
```

```python
# 묶은 것의 값을 보자.
data.groupby(['month']).groups.values()
```

```python
data.groupby(['month']).groups["2014-11"]
```

```python
len(data.groupby(['month']).groups['2014-11'])
```

```python
# 각 데이터를 보는 방법 --: 교재 350참고
for name, group in data.groupby('month'):
    print(name)
    print(group)
```

```python
# 묶은 후에 뭔가 적용하기!!!
# 달 별로 앞에 데이터들 보기!!!
data.groupby('month').head()
```

```python
# 달별로 가장 앞에 있는 데이터만 보기
data.groupby('month').first()
```

```python
# 달별로 데이터를 묶은 후에 mean하기(여기서는 할 수 있는 데이터가 duration 만 있어서...)
data.groupby('month').mean()
```

```python
# 달별로 데이터를 묶은 후에 합히기(여기서는 할 수 있는 데이터가 duration 만 이니;;)
data.groupby('month').sum()
```

```python
# 직접 묶은 후에 연산을 적용할 컬럼을 지정하고 동작 수행
data.groupby('month')['duration'].sum()
```

```python
data["month"].value_counts()
```

```python
# 직접 묶은 후에 연산 수행.
#data.groupby('month')['date'].count()
data.groupby('month').count()
```

```python
len(data[data["item"]=="call"])
```

```python
# item중에서 call인 데이터 중에서 network 로 묶은 후, 해당하는 지연시간에 대한 합을 구하기
data[data['item'] == 'call'].groupby('network')['duration'].sum()
```

```python
# 여러 조건으로 그룹바이
data.groupby(['month', 'item'])['date'].count()
```

```python
data.groupby(['month', 'network_type'])['date'].count()
```

```python
data.groupby(['month', 'item']).agg({'duration':"sum", 'network_type': "count", 'date': 'first'}) 
#data.groupby(['month', 'item']).agg({'duration':sum, 'network_type': count, 'date': first}) 
```

```python
# 동일한 기능을 앞에서 배운 pivot_table로....
# 피봇/crosstab과 유사한 기능.
pd.pivot_table(data, index=["month","item"], 
               aggfunc={'duration':'sum', 
                    'network_type': "count", 
                        'date': 'first'})
```

```python
pd.pivot_table(data, index=["month","item"], 
               aggfunc={'duration':sum, 
                    'network_type': count(), 
                        'date': first})
```

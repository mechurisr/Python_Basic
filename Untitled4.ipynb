```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re
```

```python
uniURL = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start=1&ref_=adv_nxt"

res = requests.get(uniURL) # urllib,request.urlopen(~~)
soup = bs(res.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
Title = soup.find_all("h3")[0].find("a").text
Title
```

```python
Year = soup.find_all("h3")[0].find_all("span")[1].text
Year
```

```python
Cert = soup.find_all("p")[0].find_all("span")[0].text
Cert
```

```python
Runtime = soup.find_all("p")[0].find_all("span")[2].text
Runtime
```

```python
Genre = soup.find_all("p")[0].find_all("span")[4].text
Genre = Genre.replace("\n","")
Genre = Genre.replace("            ","")
Genre
```
